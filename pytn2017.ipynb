# Data

The Forbes 2000 list is a ranking of the world's biggest companies, measured by sales, profits, assets and market value.

2004 companies.

- **rank**: the ranking of the company.
- **name**: the name of the company.
- **country**: a factor giving the country the company is situated in.
- **category**: a factor describing the products the company produces.
- **sales**: the amount of sales of the company in billion USD.
- **profits**: the profit of the company in billion USD.
- **assets**: the assets of the company in billion USD.
- **marketvalue**: the market value of the company in billion USD.

```python
import pandas as pd
df = pd.read_csv('http://vincentarelbundock.github.io/Rdatasets/csv/HSAUR/Forbes2000.csv')
```

```python
df.head()
```

# Bokeh

```python
import bokeh.charts as bc
import bokeh.plotting as bk
```

```python
fig = (df
         .sample(n=500)
         .pipe(bc.Scatter, "sales", "marketvalue"))
bk.show(fig)
```

# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots()
plot = ax.scatter(x='profits', y='assets', data=df, c='k', alpha=.15)
```

```python
df.plot.scatter(x='profits', y='assets', c='k', alpha=.15)
```

# Seaborn

```python
import seaborn as sns
sns.set(context='talk', style='ticks')
```

```python
sns.countplot(x='category', data=df)
sns.despine()
plt.xticks(rotation=90)
```

```python
sns.lmplot(x='profits', y='assets', data=df, fit_reg=False)
```

```python
sns.lmplot(x='rank', y='marketvalue', data=df, fit_reg=False)
```

```python
sns.lmplot(x='profits', y='marketvalue', data=df, fit_reg=False)
```
